```python
import os
from cv2 import cv2
from datetime import datetime


#assign raw path
path    = os.getcwd()+r'\raw'

#assign i
i = 0   

#iterate over path
for nfile in os.listdir(path):

    #when nfile is file and last3 name character is mp4 
    if os.path.isfile(os.path.join(path,nfile)) and nfile[-3:] == 'mp4':

        #assign video into variable
        video = cv2.VideoCapture(os.path.join(path,nfile))        

        #to compromise with error
        try:

            #while video still opened
            while(video.isOpened()):

                #assign video frame into frame variable
                _, frame = video.read()

                #when video frame are multiple of 15 (Normally 1 min video is 30fps, we want our program to pick 1 frame / 0.25s)
                if i % 15 == 0:

                    #assign new path of image dataset
                    time    = datetime.timestamp(datetime.now())
                    npath   = f'{os.getcwd()}\dataset\{nfile[:-4]}\{nfile[:-4]}\{nfile[:-4]}{time}.jpg'

                    #assign image height and width
                    height, width, _ = frame.shape                

                    #slicing the image                                    
                    frame  = frame[0 : int(height), int(width * (3/12)) : int(width * (9/12))]

                    #resize image into 200 x 200
                    frame  = cv2.resize(frame, (200,200))                             

                    #write image
                    cv2.imwrite(npath, frame)

                    #increment i variable                
                    i += 1           

                #when video frame are not multiple of 3
                else:   

                    #increment i variable                 
                    i +=1 
                    
        #when error is raise                     
        except:
            #pass the error & keep program run
            pass       
```

```python
from keras.preprocessing.image import ImageDataGenerator

#Generate image for more dataset
def generate(name, path):    
    image = ImageDataGenerator(
            featurewise_center= True,
            samplewise_center= True,
            featurewise_std_normalization= True,
            samplewise_std_normalization= True,
            rotation_range= 5,
            width_shift_range= [-50, 50],
            height_shift_range= [-50, 50],
            brightness_range = [0.5, 1.5],
            zoom_range= [0.5, 1],
            horizontal_flip = True,
            fill_mode = 'nearest',
            )

    #Flow to directory, the instance number are number of imagedatagenerator param * 50
    for i, batch in enumerate(image.flow_from_directory(
                directory   = path,
                save_to_dir = f'datasetready\{name}',
                target_size = (200,200),
                shuffle     = True,
                save_prefix = name)):

        if i > 50:
            break
```

```python
from datetime import datetime

#rename image
def rename(name, dir):    
    for i in os.listdir(dir):
        temp    = i[-4:]
        time    = datetime.timestamp(datetime.now())
        os.replace(os.path.join(dir,i), os.path.join(dir,f'{name}{time}{temp}'))
```

```python
generate('cahyo','dataset\cahyo')
```
